```python
import pandas as pd
import os
```

```python
geodb=pd.read_csv("geodb.csv")
geodb.columns=[x.strip() for x in list(geodb.columns)] # some fields have trailing blanks
geodb=geodb.drop(["GlobalID","CreationDate",'EditDate', 'Editor','Contact Information','Please describe the weather conditions during the flooding event.',
       'Other - Please describe the weather conditions during the flooding event.','Is the location flooded at the time of the report?',
       'May we contact you?',],axis=1)

geodb
```

```python
geodb["photo_Date"]=geodb["Date of Observation"].str.split(" ").str[0]
geodb["guessSource"]=geodb["Which of the following were flooded?"]+". "+geodb["Other - Which of the following were flooded?"]
geodb["guessCause"]=geodb["What caused the flooding?"]+". "+geodb["Other - What caused the flooding?"]
```

```python
geodb_mc=geodb.drop(["Date of Observation","Which of the following were flooded?","Other - Which of the following were flooded?",\
                     "What caused the flooding?","Other - What caused the flooding?"],axis=1)

geodb_mc.columns=["objectID","creator","post_comment","est_depth","photo_time","alternative_loc","location_longitude","location_latitude",\
                  "photo_date","guessSource","guessCause" ]

geodb_mc["image1"]=""
```
